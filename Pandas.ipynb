#**Pandas**
Pandas is an essential open-source Python library for data manipulation and analysis, created by Wes McKinney in 2008. It offers powerful data structures like Series and DataFrame, making it incredibly easy to clean, transform, and analyze data. Whether you’re dealing with simple data tasks or complex data science projects, Pandas provides the tools you need to make data-driven decisions efficiently.

```python
import pandas as pd
```

The functions *read_csv* and *read_excel* are fundamental in the Pandas library, serving as essential tools for importing data into Python. *read_csv* is used to import data from CSV (Comma-Separated Values) files, while *read_excel* is utilized for reading data from Excel files, converting them into Pandas DataFrame format.

*pd.crosstab()* function gives the frequency table that is a cross-tabulation of two variables

```python
pd.crosstab(d['league_rank'], d['international_reputation'])
```

```python
pd.crosstab(d['league_rank'], d['international_reputation'],
            margins = True,
            margins_name="Total",
            normalize = True)
```

*pd.qcut()* function bins the data or segments the data based on the distribution of the data

```python
pd.qcut(d['value_eur'], q = 5)
```

We can use the *value_counts()* on the above line of code to see how players fall in which range

```python
pd.qcut(d['value_eur'], q = 5).value_counts()
```

*pd.cut()* function is another method for binning

```python
pd.cut(d['value_eur'], bins = 5).value_counts()
```
